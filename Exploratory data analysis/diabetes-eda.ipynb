### Histogram for each variable

```python
df.hist(edgecolor="black", linewidth=1.2, figsize =(12,12));
```

### Pairplot for each variable

```python
data = df[['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DPF', 'Age', 'Outcome']]
plt.figure(figsize=(10, 10))
sns.pairplot(data,diag_kind="hist");
```

### Correlation Matrix

```python
f, ax = plt.subplots(figsize=(15, 12))
hm = sns.heatmap(round(df.corr(),2), annot=True, ax=ax, cmap="crest")
plt.title('Pearson Correlation Heatmap', fontsize = 15);
```

There is a good correlation between glucose and outcome. BMI is fairly correlated with outcome.
